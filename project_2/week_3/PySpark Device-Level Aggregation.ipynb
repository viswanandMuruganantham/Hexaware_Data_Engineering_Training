```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder \
    .appName("SmartEnergyAnalysis") \
    .getOrCreate()

# Sample Data
data = [
    ("Air Conditioner", "Peak", 5.5),
    ("Refrigerator", "Off-Peak", 2.1),
    ("LED TV", "Peak", 1.2),
    ("Fan", "Off-Peak", 0.8),
    ("Air Conditioner", "Peak", 6.5)
]

columns = [
    "device_name",
    "usage_type",
    "energy_kwh"
]

df = spark.createDataFrame(data, columns)

print("Original Data:")
df.show()

# Aggregate device usage
device_usage = df.groupBy(
    "device_name"
).agg(
    sum("energy_kwh").alias("total_usage")
)

print("Device Usage:")
device_usage.show()

# Top consuming devices
top_devices = device_usage.orderBy(
    col("total_usage").desc()
)

print("Top Energy Devices:")
top_devices.show()

spark.stop()
```

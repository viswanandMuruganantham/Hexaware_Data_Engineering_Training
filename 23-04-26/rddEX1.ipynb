```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("redd").getOrCreate()
sc = spark.sparkContext
print("creted")
```

```python
logs = [
"Rahul login",
"Sneha login",
"Arjun login",
"Rahul purchase",
"Priya login",
"Sneha logout",
"Rahul login",
"Karan login",
"Arjun purchase",
"Sneha login",
"Rahul logout",
"Amit login",
"Priya purchase",
"Karan purchase",
"Sneha login",
"Rahul purchase",
"Meera login",
"Arjun login",
"Sneha purchase",
"Rahul login"
]
log_rdd = sc.parallelize(logs)
log_rdd.collect()
```

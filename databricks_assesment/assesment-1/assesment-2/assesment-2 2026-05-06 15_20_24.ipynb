```python
from pyspark.sql.window import Window
product_rev = final_df.groupBy(
    "category",
    "product_name"
).agg(
    sum("bill_amount").alias("revenue")
)

window_cat = Window.partitionBy(
    "category"
).orderBy(desc("revenue"))

product_rev.withColumn(
    "rank",
    rank().over(window_cat)
).show()
```

```python
supplier_rev = final_df.groupBy(
    "supplier_city",
    "supplier_name"
).agg(
    sum("bill_amount").alias("revenue")
)

window_supplier = Window.partitionBy(
    "supplier_city"
).orderBy(desc("revenue"))

supplier_rev.withColumn(
    "rank",
    rank().over(window_supplier)
).show()
```

```python
window_top = Window.partitionBy(
    "category"
).orderBy(desc("revenue"))

product_rev.withColumn(
    "row_num",
    row_number().over(window_top)
).filter(
    col("row_num") == 1
).show()
```

```python
window_dense = Window.orderBy(desc("bill_amount"))

payments_df.withColumn(
    "dense_rank",
    dense_rank().over(window_dense)
).show()
```

```python
supplier_total = final_df.groupBy(
    "supplier_name"
).agg(
    sum("bill_amount").alias("revenue")
)

window_sup = Window.orderBy(desc("revenue"))

supplier_total.withColumn(
    "rank",
    dense_rank().over(window_sup)
).filter(
    col("rank") <= 2
).show()
```

```python
daily_rev = final_df.groupBy(
    "order_date"
).agg(
    sum("bill_amount").alias("daily_revenue")
)

window_run = Window.orderBy("order_date")

daily_rev.withColumn(
    "running_total",
    sum("daily_revenue").over(window_run)
).show()
```

```python
daily_orders_df.createOrReplaceTempView(
    "daily_updates"
)
```

```python
%sql
MERGE INTO orders_target t
USING daily_updates s
ON t.order_id = s.order_id

WHEN MATCHED THEN
UPDATE SET
  t.order_status = s.order_status,
  t.quantity = s.quantity

WHEN NOT MATCHED THEN
INSERT (order_id, product_id, supplier_id, order_date, quantity, order_status)
VALUES (s.order_id, s.product_id, s.supplier_id, s.order_date, s.quantity, s.order_status)
```

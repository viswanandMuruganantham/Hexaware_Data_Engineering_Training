```python
import ipywidgets as widgets
from IPython.display import display
```

```python
name_widgets = widgets.Text(
    value = "enter name",
    description = "Name:"
)
display(name_widgets)
```

```python
dept_widgets = widgets.Dropdown(
    options = ["IT", "HR", "Finance"],
    value = "IT",
    description = "Dept"
)
display(dept_widgets)
```

```python
salary_slider = widgets.IntSlider(
    value = 60000,
    min = 30000,
    max = 100000,
    steps = 5000,
    description = "salary"
)
display(salary_slider)
```

```python
button = widgets.Button(description = "Show values")
def on_button(b):
  print("NAme",name_widgets.value)
  print("dept", dept_widgets.value)
  print("salary_slider", salary_slider.value)
button.on_click(on_button)
display(button)
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("combined").getOrCreate()
print("created")
```

```python
employees1 = spark.read.csv("employees1.csv", header = True, inferSchema=True)
employees1.show()
```

```python
logins = spark.read.text("logins.txt")
logins.show()
```

```python
from pyspark.sql.functions import explode
odf= spark.read.option("multiLine", True).json("orders1.json")
orders = odf.select(explode("orders")).select("col.*")
orders.show(truncate=False)
```

```python
sales = spark.read.csv("sales1.csv", header = True, inferSchema = True)
sales.show()
```

```python
logins.show()
```

```python
logins.count()
```

```python
logins.select("value").distinct().show()
```

```python
count_log = logins.groupBy("value").count()
count_log.show()
```

```python
from pyspark.sql.functions import col
count_log.orderBy(col("count").desc()).show(3)
```

```python
count_log.filter(col("count") > 4).show()
```

```python
result = {row["value"]: row["count"] for row in count_log.collect()}
print(result)
```

```python
sales.show()
```

```python
rev=sales.withColumn("Revenue", col("quantity")*col("price"))
rev.show()
```

```python
from pyspark.sql.functions import sum
rev.agg(sum("Revenue")).show()
```

```python
rev.groupBy("product").agg(sum("Revenue").alias("pre_pro")).show()
```

```python
qua = rev.groupBy("product").agg(sum("quantity").alias("sum_quantity"))
qua.show()
```

```python
qua.orderBy(col("sum_quantity").desc()).show(1)
```

```python
rev.groupBy("emp_id").agg(sum("Revenue").alias("sum_revenue")).orderBy(col("sum_revenue").desc()).show(1)
```

```python
rev.agg(avg("Revenue").alias("avg_Revenue")).show()
```

```python
rev.groupBy("product").agg(sum("Revenue").alias("sum_rev")).filter(col("sum_rev") > 100000).show()
```

```python
df = employees1.join(sales, on = "emp_id", how = "inner")
df.show()
```

```python
df = employees1.join(rev, on ="emp_id", how = "inner")
df.show()
```

```python
sal = df.groupBy("emp_id", "name").agg(sum("Revenue").alias("su"))
sal.show()
```

```python
sal.orderBy(col("su").desc()).show(5)
```

```python
df.groupBy("department").agg(sum("revenue").alias("sum_rev")).orderBy(col("sum_rev").desc()).show(1)
```

```python
from pyspark.sql.functions import col, sum, rank
from pyspark.sql.window import Window

df = orders.join(customers, "customer_id") \
           .join(order_items, "order_id") \
           .join(products, "product_id") \
           .withColumn("revenue", col("quantity") * col("price")) \
           .groupBy("customer_id", "name") \
           .agg(sum("revenue").alias("total_revenue"))

window_spec = Window.orderBy(col("total_revenue").desc())

df.withColumn("rank", rank().over(window_spec)).show()
```

```python
from pyspark.sql.functions import col, sum, rank
from pyspark.sql.window import Window

df = order_items.join(products, "product_id") \
                .withColumn("revenue", col("quantity") * col("price")) \
                .groupBy("product_id", "product_name", "category") \
                .agg(sum("revenue").alias("total_revenue"))


window_spec = Window.partitionBy("category").orderBy(col("total_revenue").desc())

df.withColumn("rank", rank().over(window_spec)) \
  .filter(col("rank") == 1) \
  .show()
```

```python
from pyspark.sql.functions import col, sum
from pyspark.sql.window import Window


df = orders.join(order_items, "order_id") \
           .join(products, "product_id") \
           .withColumn("revenue", col("quantity") * col("price")) \
           .groupBy("order_id", "order_date") \
           .agg(sum("revenue").alias("order_revenue"))


window_spec = Window.orderBy("order_date") \
                    .rowsBetween(Window.unboundedPreceding, Window.currentRow)

df.withColumn("running_total", sum("order_revenue").over(window_spec)) \
  .orderBy("order_date") \
  .show()
```

```python
from pyspark.sql.functions import col, sum
from pyspark.sql.window import Window

df = orders.join(order_items, "order_id") \
           .join(products, "product_id") \
           .withColumn("revenue", col("quantity") * col("price"))


window_spec = Window.partitionBy("product_id") \
                    .orderBy("order_date") \
                    .rowsBetween(Window.unboundedPreceding, Window.currentRow)

df.withColumn("running_total", sum("revenue").over(window_spec)) \
  .select("product_id", "product_name", "order_date", "revenue", "running_total") \
  .orderBy("product_id", "order_date") \
  .show()
```

```python
spark.sql("""
SELECT c.customer_id, c.name,
       SUM(oi.quantity * p.price) AS total_revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.customer_id, c.name
ORDER BY total_revenue DESC
LIMIT 3
""").show()
```

```python
spark.sql("""
SELECT p.product_id, p.product_name,
       SUM(o.quantity * p.price) AS total_revenue
FROM order_items o
JOIN products p
ON o.product_id = p.product_id
GROUP BY p.product_id, p.product_name
ORDER BY total_revenue DESC
LIMIT 1
""").show()
```

```python
final_df = spark.sql("""
SELECT
    o.order_id,
    c.customer_id,
    c.name,
    c.city,
    p.product_id,
    p.product_name,
    p.category,
    oi.quantity,
    p.price,
    (oi.quantity * p.price) AS revenue,
    o.order_date
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
""")

final_df.show()
```

```python
spark.sql("""
SELECT *,
       RANK() OVER (ORDER BY total_revenue DESC) AS rank
FROM (
    SELECT c.customer_id, c.name,
           SUM(oi.quantity * p.price) AS total_revenue
    FROM customers c
    JOIN orders o ON c.customer_id = o.customer_id
    JOIN order_items oi ON o.order_id = oi.order_id
    JOIN products p ON oi.product_id = p.product_id
    GROUP BY c.customer_id, c.name
) t
""").show()
```

````python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL Lab").getOrCreate()


employees_data = [
    (1, "Rahul", "IT", 70000, "Hyderabad"),
    (2, "Sneha", "HR", 60000, "Bangalore"),
    (3, "Arjun", "IT", 75000, "Chennai"),
    (4, "Priya", "Finance", 80000, "Hyderabad"),
    (5, "Karan", "IT", 50000, "Mumbai"),
    (6, "Amit", "HR", 58000, "Delhi"),
    (7, "Meera", "Finance", 82000, "Bangalore")
]

employees_cols = ["emp_id", "name", "department", "salary", "city"]

employees_df = spark.createDataFrame(employees_data, employees_cols)
employees_df.show()


departments_data = [
    ("IT", "Technology"),
    ("HR", "People Operations"),
    ("Finance", "Accounts and Finance")
]

departments_cols = ["department", "dept_full_name"]

departments_df = spark.createDataFrame(departments_data, departments_cols)
departments_df.show()

sales_data = [
    (101, 1, "Laptop", 1, 75000),
    (102, 2, "Mouse", 3, 500),
    (103, 3, "Keyboard", 2, 1500),
    (104, 1, "Monitor", 1, 12000),
    (105, 4, "Laptop", 1, 75000),
    (106, 3, "Mouse", 2, 500),
    (107, 5, "Keyboard", 1, 1500),
    (108, 1, "Laptop", 1, 75000)
]

sales_cols = ["sale_id", "emp_id", "product", "quantity", "price"]
sales_df = spark.createDataFrame(sales_data, sales_cols)
sales_df.show()
````

````python
employees_df.createOrReplaceTempView("employees")
departments_df.createOrReplaceTempView("depart")
sales_df.createOrReplaceTempView("sales")
````

````python
spark.sql("select * from employees").show()
````

Great — this line is **very important when using Spark SQL**. Let’s break it down clearly.

---

# 🔥 Code

```python
employees_df.createOrReplaceTempView("employees")
```

---

# 🧠 What it does (Simple Meaning)

👉 It **registers your DataFrame as a temporary SQL table (view)**

So after this:

```text
employees_df  →  "employees" (SQL table)
```

---

# 🔹 Why we use it?

👉 To run **SQL queries** on DataFrame

---

# 🔥 Example Flow

## Step 1: Create DataFrame

```python
employees_df.show()
```

---

## Step 2: Register as SQL view

```python
employees_df.createOrReplaceTempView("employees")
```

---

## Step 3: Use SQL

```python
spark.sql("SELECT * FROM employees").show()
```

---

# 🔍 Breakdown of the name

### ✔ `create`

👉 Create a view

### ✔ `OrReplace`

👉 If already exists → replace it

### ✔ `TempView`

👉 Temporary (session-level)

---

# 🔥 What is a "Temp View"?

👉 It is:

* Stored in **Spark session memory**
* Not saved permanently
* Exists only while session is running

---

# ⚠️ Important Rules

### ❌ Not persistent

* If Spark session stops → view disappears

---

### ❌ Not stored in database

* Not in MySQL / Hive (unless global or permanent)

---

# 🔹 Types of Views

| Type             | Scope           | Lifetime             |
| ---------------- | --------------- | -------------------- |
| Temp View        | Current session | Until session ends   |
| Global Temp View | Across sessions | Until Spark app ends |

---

## 🔥 Global View Example

```python
employees_df.createOrReplaceGlobalTempView("employees")
```

Query:

```python
spark.sql("SELECT * FROM global_temp.employees").show()
```

---

# 🧠 Real-world Use Case

👉 You use this when:

* Mixing **SQL + PySpark**
* Analysts prefer SQL
* Complex queries (joins, aggregations)

---

# 🔥 Example Query

```python
spark.sql("""
SELECT department, AVG(salary)
FROM employees
GROUP BY department
""").show()
```

---

# 🧠 Simple Analogy

👉 Think like:

```text
DataFrame → Excel Sheet
Temp View → SQL Table Name
```

````python
spark.sql("""
select count(*) as total_emp, sum(salary) as total_salary,
avg(salary) as avg_salary, max(salary) as Max_salary,
min(salary) as min_salary from employees """).show()
````

````python
spark.sql("""
select department, avg(salary) as Avg from employees group by department
""").show()
````

````python
spark.sql("""
select department, count(*) as emp_count
from employees group by department """).show()
````

````python
spark.sql("""
select e.name, sum(s.quantity * s.price) as total_sales
from employees e join sales s on e.emp_id = s.emp_id
group by e.name """).show()
````

````python
spark.sql("""
select  department, avg(salary) as sal from employees
group by department
having avg(salary) > 70000 """).show()
````

````python
spark.sql("""
select product, sum(quantity) as qu from sales
group by product having sum(quantity) > 2 """).show()
````

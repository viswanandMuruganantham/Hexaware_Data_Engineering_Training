```python
from pyspark.sql.window import Window
```

```python
cus_acc = customers.join(accounts, on="customer_id")
cus_acc.show()
```

```python
windo = Window.partitionBy("city").orderBy(col("balance").desc())
cus_acc.withColumn("rank", rank().over(windo)).show()
```

```python
cus_acc.withColumn(
    "rn",
    row_number().over(windo)
).filter("rn = 1").show()
```

```python
wind_tx = Window.partitionBy("txn_type").orderBy(col("amount").desc())
transactions.withColumn("rank", dense_rank().over(wind_tx)).show()
```

```python
branches.show()
accounts.show()
```

```python
branch_bal = accounts.groupBy("branch") \
                     .agg(sum("balance").alias("total_balance"))
window_spec = Window.orderBy(col("total_balance").desc())

result = branch_bal.withColumn("rank", rank().over(window_spec))
result.show()
```

```python
final_df = customers.join(accounts, "customer_id") \
                    .join(transactions, "account_id")
cust_txn = final_df.groupBy("name").agg(sum("amount").alias("total_txn"))

cust_txn.withColumn(
    "rank",
    rank().over(Window.orderBy(col("total_txn").desc()))
).show()
```

```python
cust_city_txn = final_df.groupBy("name", "city") \
                       .agg(sum("amount").alias("total_txn"))

window_city = Window.partitionBy("city").orderBy(col("total_txn").desc())

cust_city_txn.withColumn(
    "rn",
    row_number().over(window_city)
).filter("rn = 1").show()
```

```python
cust_full = customers.join(accounts, "customer_id") \
                     .join(branches, "branch")

window_region = Window.partitionBy("region").orderBy(col("balance").desc())

cust_full.withColumn(
    "rn",
    row_number().over(window_region)
).filter("rn = 1").show()
```

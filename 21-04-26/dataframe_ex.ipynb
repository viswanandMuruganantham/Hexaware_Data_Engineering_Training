```python
# Exercise 31
employees.withColumn("bonus", col("salary") * 0.1).show()

# Exercise 32
employees.withColumn("company", col("salary") * 0 + 1) \
    .withColumn("company", col("company") * 0 + 1) \
    .withColumn("company", col("company") * 0 + 1) \
    .withColumn("company", col("company") * 0 + 1) \
    .withColumn("company", col("company") * 0 + 1)

# Better:
from pyspark.sql.functions import lit
employees.withColumn("company", lit("BotCampus")).show()
```

```python
# Exercise 33
employees.filter(col("department").isNull()).show()

# Exercise 34
employees.fillna({"department": "Unknown"}).show()

# Exercise 35
employees.dropna().show()
```

```python
from pyspark.sql.functions import upper, length

# Exercise 36
employees.withColumn("name_upper", upper("name")).show()

# Exercise 37
employees.withColumn("name_length", length("name")).show()

# Exercise 38
employees.withColumn("salary_in_lakhs", col("salary") / 100000).show()
```

```python
from pyspark.sql.functions import when

# Exercise 39
employees.withColumn(
    "salary_grade",
    when(col("salary") >= 75000, "High")
    .when(col("salary") >= 60000, "Medium")
    .otherwise("Low")
).show()
```

```python
from pyspark.sql.functions import to_date, year, month

# Exercise 40
employees = employees.withColumn("joining_date", to_date("joining_date"))

# Exercise 41
employees.withColumn("year", year("joining_date")) \
         .withColumn("month", month("joining_date")).show()
```

```python
new_data = spark.createDataFrame([
    (8, "Ravi", "IT", 65000, "2023-05-01"),
    (9, "Anu", "HR", 62000, "2023-06-01")
], employees.columns)

# Exercise 42
employees.union(new_data).show()

# Exercise 43
employees.unionByName(new_data).show()
```

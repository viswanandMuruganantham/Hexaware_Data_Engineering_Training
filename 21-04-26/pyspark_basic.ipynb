```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("basics_pyspark").getOrCreate()
print("session created")
```

```python
print(spark.version)
```

```python
data = [
    ("viswa", 21, "tvr"),
    ("murug", 22, "str"),
    ("siva", 23, "ret")
]
col = ["name", "age", "city"]
df = spark.createDataFrame(data,col)
df.show()
```

```python
df.printSchema()
```

```python
df.select("Name").show()
```

```python
df.filter(df.age>21).show()
```

```python
df.groupby("city").count().show()
```

```python
df.orderBy("age").show()
```

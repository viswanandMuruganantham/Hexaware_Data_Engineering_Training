```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ex2").getOrCreate()
data = [
    (1, "Rahul", "IT", 70000, "2024-01-10"),
    (2, "Sneha", "HR", 60000, "2024-02-15"),
    (3, "Arjun", "IT", 75000, "2024-03-01"),
    (4, "Priya", "Finance", 80000, "2024-01-25"),
    (5, "Karan", None, 50000, "2024-02-20")
]

columns = ["emp_id", "name", "department", "salary", "joining_date"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
df.show()
df.printSchema()
df.columns
df.count()
```

```python
df.show(3)
```

```python
df_rename = df.withColumnRenamed("salary", "emp_salary")
df_rename.show()
```

```python
df.filter(df.department == "IT").show()
```

```python
df.drop("joining_date").show()
```

```python
from pyspark.sql.functions import sum, avg, min, max
df.select(sum("salary").alias("total_sal"), avg("salary").alias("avg_sal"), max("salary").alias("max_sal"), min("salary")).show()
```

```python
dept_data = [
    ("IT", "Bangalore"),
    ("HR", "Mumbai"),
    ("Finance", "Delhi")
]

dept_columns = ["department", "location"]

dept_df = spark.createDataFrame(dept_data, dept_columns)
dept_df.show()
```

```python
df.join(dept_df, on = "department" , how = "inner").show()
```

```python
df.join(dept_df, on = "department", how = "left").show()
```

```python
df.join(dept_df, on = "department", how = "right").show()
```

```python
df.withColumn("bonus", df.salary * 0.10).show()
```

```python
df.show()
```

```python
from pyspark.sql.functions import col
df.filter(col("department").isNull()).show()
```

```python
from pyspark.sql.functions import upper, lower, length
df.select(upper("name").alias("upper_case"), lower("name").alias("lower_case"), length("name").alias("len")).show()
```
